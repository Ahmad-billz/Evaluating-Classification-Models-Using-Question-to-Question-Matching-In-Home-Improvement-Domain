```python
with open('/content/drive/My Drive/thesis/dataset_building/data_updated.json') as d_file:
    data_new= json.load(d_file)
df = pd.DataFrame.from_dict(data_new, orient='index')
```

```python
df.head(25)
```

```python
dic_file={  
            'q1_id': [],
            'q2_id': [],
            'q1': [],
            'c1': [],
            'q2': [],
            'c2': []}
```

```python
df = df.replace({r'\s+$': '', r'^\s+': ''}, regex=True).replace(r'\n',  ' ', regex=True)
```

```python
df.index
```

```python
for ind in df.index:
  print(df.index)
```

```python
for ind in df.index:
  i=df.index.get_loc(ind)
  q=df.iloc[i]['question']
  c=df.iloc[i]['context']
  for ind2 in df.iloc[i]['linked_questions']:
    if ind2 in df.index: #as dataset is smaller so does not have all question in it , actually used this if to avoid error
      j = df.index.get_loc(ind2)
      q_r = df.iloc[j]['question']
      c_r = df.iloc[j]['context']
      #print(q+"\n"+q_r+"\n")
      print(ind+"\n"+ind2+"\n")
      #q_doc2=q_r+"\n"+c_r
      
      dic_file['q1_id'].append(ind)
      dic_file['q2_id'].append(ind2)
      dic_file['q1'].append(q)
      dic_file['c1'].append(c)
      dic_file['q2'].append(q_r)
      dic_file['c2'].append(c_r)
```

```python
df2 = pd.DataFrame(dic_file)
```
